# Preparando o Ambiente

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/didi-a-p/BootcampDataScienceAlura/main/Analisando%20dados%20da%20sa%C3%BAde%20com%20Python%20e%20Pandas/dados/A151346189_28_143_208.csv"
```

```python
# dados = pd.read_csv(url, sep=";", skiprows=3, skipfooter=12, encoding="cp1252", engine='python',thousands='.', decimal=',')
```

```python
dados = pd.read_csv(url, sep=";", skiprows=3, skipfooter=12, encoding="ISO-8859-1",engine='python',thousands='.', decimal=',')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.ticker as ticker
```

```python
ax =  dados.plot(x='Unidade da Federação', y='2008/Ago', kind='bar', figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}")) #"{:.2f}"
plt.title("Valor por unidade da Federação")
plt.show()
```
